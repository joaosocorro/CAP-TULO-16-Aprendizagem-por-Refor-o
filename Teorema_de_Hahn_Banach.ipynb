```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for x
x = np.linspace(0, 1, 400)

# Define the function alpha * x^2
def func(alpha):
    return alpha * x**2

# Calculate the values for different alphas
alphas = np.linspace(-1, 1, 5)
funcs = [func(alpha) for alpha in alphas]
g1_values = [alpha * (1/3) for alpha in alphas]
g3_values = [alpha for alpha in alphas]

# Plot for g1
plt.figure(figsize=(12, 6))
for alpha, f, g1_val in zip(alphas, funcs, g1_values):
    plt.plot(x, f, label=f'alpha={alpha}, integral = {g1_val:.2f}')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('Extenso $\\widetilde{g}_1$ com integral de 0 a 1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()

# Plot for g3
plt.figure(figsize=(12, 6))
for alpha, f, g3_val in zip(alphas, funcs, g3_values):
    plt.plot(x, f, label=f'alpha={alpha}, f(1) = {g3_val:.2f}')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('Extenso $\\widetilde{g}_3$ com $f(1)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```
